```python
import pandas as pd

df = pd.read_csv("../data/cleaned_customer_data.csv")
```

```python
# Tenure segments
df['TenureSegment'] = pd.cut(
    df['tenure'],
    bins=[0, 12, 36, 72],
    labels=['New', 'Mid-Term', 'Long-Term']
)
```

```python
# Churn by tenure segment
pd.crosstab(df['TenureSegment'], df['Churn'], normalize='index') * 100
```

```python
# Contract + tenure segmentation
segment_summary = df.groupby(
    ['TenureSegment', 'Contract']
)['ChurnFlag'].mean().reset_index()

segment_summary
```

```python
# High risk customer flag
df['HighRiskCustomer'] = (
    (df['TenureSegment'] == 'New') &
    (df['Contract'] == 'Month-to-month')
)

df['HighRiskCustomer'].value_counts(normalize=True) * 100
```
