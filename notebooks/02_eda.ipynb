```python
import pandas as pd

df = pd.read_csv("../data/cleaned_customer_data.csv")
```

```python
# Overall churn rate
overall_churn = df['ChurnFlag'].mean()
print(f"Overall Churn Rate: {round(overall_churn*100,2)}%")
```

```python
# Churn by gender
pd.crosstab(df['gender'], df['Churn'], normalize='index') * 100
```

```python
# Churn by senior citizen
pd.crosstab(df['SeniorCitizen'], df['Churn'], normalize='index') * 100
```

```python
# Tenure vs churn
df.groupby('ChurnFlag')['tenure'].mean()
```

```python
# Monthly charges vs churn
df.groupby('ChurnFlag')['MonthlyCharges'].mean()
```

```python
# Contract type vs churn
pd.crosstab(df['Contract'], df['Churn'], normalize='index') * 100
```

```python
# Payment method vs churn
pd.crosstab(df['PaymentMethod'], df['Churn'], normalize='index') * 100
```
